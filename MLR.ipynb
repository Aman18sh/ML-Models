# Implementing Multi Linear Regression on Numeric Data 

As per my opinion, there is no need of considering categorical columns while implementing linear regression, but if there is a target column like prices then i can consider categorical columns for doing price prediction but for now i m just considering numeric part to implement multi-linear regression.

```python
data_int
```

# Splitting data in to target and features

```python
x = data_int[['Dimensions.Height', 'Dimensions.Length', 'Dimensions.Width',
       'Engine Information.Number of Forward Gears',
       'Fuel Information.City mpg', 'Fuel Information.Highway mpg',
       'Identification.Year',
       'Engine Information.Engine Statistics.Torque']]
y = data_int['Engine Information.Engine Statistics.Horsepower'].values.reshape(x.shape[0],1)
y
```

```python
x
```

```python
y
```

# Multi Linear Regression With Outliers

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
# splitting the data for training and testing

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=2)
x_train.shape,x_test.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr =  LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr.intercept_
```

```python
lr.coef_
```

```python
y_pred = lr.predict(x_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,y_pred)
```
